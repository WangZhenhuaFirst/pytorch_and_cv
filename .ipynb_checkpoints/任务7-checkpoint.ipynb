## 任务7：PyTorch池化层和归一化层

```python
# 使用pytroch 实现 2d pool中的 mean-pooling

import torch

m = torch.nn.AvgPool2d(3, stride=1)
input_avg = torch.randn(3, 5, 5)
print("input_avg:", input_avg)
output_avg = m(input_avg)
print("output_avg:", output_avg)
```

```python
# 使用pytroch 实现 2d pool中的 max-pooling

import torch

m = torch.nn.MaxPool2d(3, stride=1)
input_max = torch.randn(3, 5, 5)
print('input_max:', input_max)
output_max = m(input_max)
print('output_max:', output_max)
```
